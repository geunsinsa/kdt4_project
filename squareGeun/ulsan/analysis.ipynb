```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import utill
import statsmodels.api as sm
ulsanDF = pd.read_csv('ulsan.csv').drop(columns='Unnamed: 0')
ulsanDF['ulsan_avg_temp'].fillna(method='ffill',inplace=True)

model2 = sm.OLS(ulsanDF['precipitation(mm)'],sm.add_constant(ulsanDF['ulsan_avg_temp']))
print(model2.fit().summary())
```
