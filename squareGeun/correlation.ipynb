```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import utill

utill.set_os_specific_font()
```

```python
df = pd.read_csv('water_temp.csv').drop(columns='Unnamed: 0')
```

```python
sns.set_style('dark')
plt.figure(figsize=(10,10))
sns.lineplot(data=df, x='연도',y='동해평균수온',label='East')
sns.lineplot(data=df, x='연도',y='남해평균수온',label='South')
sns.lineplot(data=df, x='연도',y='서해평균수온',label='West')
sns.lineplot(data=df, x='연도',y='전체평균수온',label='Total')
plt.ylabel('temperature',size=15)
plt.yticks(size=10)
plt.xlabel('Year',size=15)
plt.xticks(size=10)
plt.legend()
plt.grid(True)
plt.savefig('a.png')
```

```python
sns.set_style('dark')
plt.figure(figsize=(10,10))
sns.lmplot(data=df, x='연도',y='전체평균수온')
plt.ylabel('temperature',size=15)
plt.yticks(size=10)
plt.xlabel('Year',size=15)
plt.xticks(size=10)
plt.legend()
plt.grid(True)
plt.savefig('a1.png')
```
